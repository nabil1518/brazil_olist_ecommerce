## The Final Directive

### Affects of delivery delays

Delivery delays are eroding our customer experience. Our analysis reveals a direct link between late deliveries and dissatisfaction: orders arriving early earn an average 4.3-star review, while on-time drops to 4.0, and late orders fall further. Crucially, the first delivery experience shapes long-term loyalty—customers whose initial order is on-time return at a 3.1% repeat purchase rate, compared to just 2.6% for those who experience a delay. The supply chain also shows troubling volatility, with monthly failure rates swinging wildly and a spike to 100% in September 2016, underscoring an inconsistent network.

### The root cause

The root causes are physical and geographic, not preparation-time related. Seller processing speed does not meaningfully predict delays; the breakdown occurs in transit. Heavier and bulkier items are systematically more vulnerable: delayed shipments have a median weight of 800g (vs. 700g on-time) and 10% larger volume. The “audio” category suffers a 12.1% failure rate, far above the 8.3% for “health & beauty.” Geographically, the state of MA is a clear outlier—it pays the highest average freight ($30.09) yet endures the highest failure rate (20%), proving that premium shipping spend does not guarantee reliable service in all regions.

### The Impact 

Lower review scores depress conversion and search rank, while the repeat-purchase gap translates to a meaningful lifetime-value loss. With thousands of customers affected, even a half-point retention lift can drive significant incremental revenue. Additionally, our risk-scoring exercise shows that shipments flagged with three risk factors (excess weight, volume, and freight cost) fail at 7.8% versus 6.1% for those with no risk, highlighting a clear segment where intervention yields the highest return.

### Recommendations

Our recommendation is a three-point action plan. First, operationalize a risk-flagging system at order placement: automatically identify shipments that exceed weight/volume thresholds or are bound for high-failure states like MA, and route them to premium carriers with proven reliability in those corridors—not simply the most expensive option. Second, work with carriers to create specialized handling for the “audio” and other heavy/bulky categories, potentially using dedicated regional sortation. Third, launch a proactive communication and service-recovery protocol for flagged orders, turning a potential negative experience into a moment of trust. These targeted changes will raise review scores, strengthen repeat purchase rates, and directly improve customer lifetime value.
