## Geography & Freight Analysis

```python
# Import required libraries
import numpy as np
import pandas as pd
```

```python
# Get merged dataframe supply_chain_autopsy.
df = pd.read_csv("../data/processed/supply_chain_autopsy.csv")
```

```python
# Group by state
state_metrics = df.groupby("seller_state").agg(
    total_orders = ("order_id", "nunique"),
    total_delayed_orders = ('delivery_delay_days', lambda x: (x > 0).sum()),
    avg_freight_value = ("freight_value", "mean")
).reset_index()
state_metrics.sample(5)
```

```python
state_metrics["avg_freight_value"] = state_metrics["avg_freight_value"].round(2)

# Failure rate by state
state_metrics["failure_rate_pct"] = (
    (state_metrics["total_delayed_orders"] / state_metrics["total_orders"]) * 100).round(2)
```

```python
# States with more than 100 orders
more_than_100_orders = state_metrics[state_metrics["total_orders"] >= 100].reset_index(drop=True)
```

```python
more_than_100_orders.sort_values("failure_rate_pct", ascending=False).reset_index(drop=True).head()
```

```python
state_metrics.to_csv("../data/processed/geography_freight_analysis.csv", index=False)
```

### Observation

States with higher average freight values do not necessarily have lower failure rates—for example, MA has the highest average freight ($30.09) yet the highest failure rate (20%), indicating that paying more for shipping does not guarantee better delivery performance in certain regions.
