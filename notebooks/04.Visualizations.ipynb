## The Visual Story

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/geography_freight_analysis.csv")
```

```python
more_than_100_orders = df[df["total_orders"] >= 100].reset_index(drop=True)
```

```python
more_than_100_orders.sort_values("failure_rate_pct", ascending=False).reset_index(drop=True).head()
```

```python
# Set style for better-looking plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 7)

# Create the scatter plot with regression line
fig, ax = plt.subplots(figsize=(12, 7))

# Use regplot to show the trend line with confidence interval
sns.regplot(
    data=more_than_100_orders,
    x='avg_freight_value',
    y='failure_rate_pct',
    scatter_kws={'s': 100, 'alpha': 0.6, 'color': 'steelblue'},
    line_kws={'color': 'red', 'linewidth': 2, 'label': 'Regression Line'},
    ci=95,
    ax=ax
)

# Find the data point for "MA"
ma_point = more_than_100_orders[more_than_100_orders['seller_state'] == 'MA']

if not ma_point.empty:
    # Get coordinates for MA
    ma_x = ma_point['avg_freight_value'].values[0]
    ma_y = ma_point['failure_rate_pct'].values[0]
    
    # Annotate the MA point
    ax.annotate(
        'MA (Anomaly)',
        xy=(ma_x, ma_y),
        xytext=(ma_x + 5, ma_y + 5),  # Offset the text slightly
        fontsize=11,
        fontweight='bold',
        color='darkred',
        bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
        arrowprops=dict(
            arrowstyle='->',
            connectionstyle='arc3,rad=0.2',
            color='darkred',
            lw=1.5
        )
    )
    
    # Highlight the MA point with a different color marker
    ax.scatter(ma_x, ma_y, s=200, color='darkred', edgecolors='black', 
               linewidth=2, zorder=5, marker='D')

# Add title and labels
ax.set_title('Shipping Cost vs. Delivery Failure Rate by State', 
             fontsize=16, fontweight='bold', pad=20)
ax.set_xlabel('Average Freight Value (USD)', fontsize=13, fontweight='medium')
ax.set_ylabel('Delivery Failure Rate (%)', fontsize=13, fontweight='medium')

# Add a text box explaining the trend
trend_note = ("Trend: Higher shipping costs are associated with\n"
              "slightly LOWER failure rates (negative correlation)\n"
              "→ Paying more generally means BETTER service in Brazil")
ax.text(0.05, 0.95, trend_note, transform=ax.transAxes,
        fontsize=10, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

# Add grid for better readability
ax.grid(True, alpha=0.3)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Display the plot
plt.show()
```

### Observation

The regression plot shows a slight negative correlation between average freight value and failure rate, higher shipping costs are actually associated with slightly lower failure rates, except for the MA anomaly which deviates significantly from the trend line.
