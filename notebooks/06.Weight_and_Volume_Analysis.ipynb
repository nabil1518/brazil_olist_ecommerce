## The Weight & Volume Analysis

```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import merged catgory_name_translated merged dataset
df = pd.read_csv("../data/processed/catgory_name_translated.csv")
pd.set_option('display.max_columns', None)
```

```python
df.dropna(subset=["product_length_cm", "product_height_cm", "product_width_cm"], inplace=True)
```

```python
df["product_volume_cm3"] = df[["product_length_cm", "product_height_cm", "product_width_cm"]].prod(axis=1)
df.iloc[:, -1:].head()
```

```python
df["is_delayed"] = df["delivery_delay_days"] > 0  
```

```python
df.groupby("is_delayed").agg(
    median_product_weight = ("product_weight_g", "median"),
    median_product_volume = ("product_volume_cm3", "median")
)
```

```python
# Set style
sns.set_style("whitegrid")

# Create figure
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot 1 - without hue
sns.boxplot(data=df, x="is_delayed", y="product_weight_g", ax=ax1, 
            color="#2ecc71", showfliers=False)  # showfliers=False removes outliers for cleaner look
ax1.set_title("Product Weight: On-Time vs Delayed Packages", fontsize=14, fontweight='bold')
ax1.set_xlabel("Delivery Status", fontsize=12)
ax1.set_ylabel("Product Weight (grams)", fontsize=12)
ax1.set_xticklabels(["On-Time", "Delayed"])

# Boxplot 2 - without hue
sns.boxplot(data=df, x="is_delayed", y="product_volume_cm3", ax=ax2,
            color="#e74c3c", showfliers=False)
ax2.set_title("Product Volume: On-Time vs Delayed Packages", fontsize=14, fontweight='bold')
ax2.set_xlabel("Delivery Status", fontsize=12)
ax2.set_ylabel("Product Volume (cm³)", fontsize=12)
ax2.set_xticklabels(["On-Time", "Delayed"])

plt.suptitle("Do Delayed Packages Have Different Physical Characteristics?", 
             fontsize=16, fontweight='bold', y=1.02)
```

### Observation

Delayed orders show higher median product weight (800g vs. 700g) and higher median product volume (7,134 cm³ vs. 6,480 cm³) compared to on-time deliveries, indicating that larger and heavier items face greater logistical challenges and are more prone to delays.
